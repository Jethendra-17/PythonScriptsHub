```python
import tkinter as tk
from tkinter import ttk

def convert_number():
    base = int(base_entry.get())
    input_number = input_entry.get()

    try:
        decimal_number = int(input_number, base)
        binary = bin(decimal_number)[2:]
        octal = oct(decimal_number)[2:]
        hexadecimal = hex(decimal_number)[2:]

        decimal_result.set(f"Decimal: {decimal_number}")
        binary_result.set(f"Binary: {binary}")
        octal_result.set(f"Octal: {octal}")
        hexadecimal_result.set(f"Hexadecimal: {hexadecimal}")
    except ValueError:
        decimal_result.set("Invalid input")
        binary_result.set("Invalid input")
        octal_result.set("Invalid input")
        hexadecimal_result.set("Invalid input")

# Create a tkinter window
root = tk.Tk()
root.title("Number System Converter")
root.geometry("400x400")  # Set window size

# Style configuration
style = ttk.Style()
style.configure("TLabel", background="#3498db", foreground="white", font=("Arial", 12))
style.configure("TButton", background="#2ecc71", foreground="black", font=("Arial", 12))
style.configure("TEntry", font=("Arial", 12))

# Base input
base_label = ttk.Label(root, text="Enter Base:")
base_label.pack(pady=10)
base_entry = ttk.Entry(root, font=("Arial", 12))
base_entry.pack(pady=10)

# Number input
input_label = ttk.Label(root, text="Enter Number:")
input_label.pack()
input_entry = ttk.Entry(root, font=("Arial", 12))
input_entry.pack(pady=10)

# Convert button
convert_button = ttk.Button(root, text="Convert", command=convert_number)
convert_button.pack(pady=20)

# Result labels
decimal_result = tk.StringVar()
decimal_label = ttk.Label(root, textvariable=decimal_result)
decimal_label.pack()

binary_result = tk.StringVar()
binary_label = ttk.Label(root, textvariable=binary_result)
binary_label.pack()

octal_result = tk.StringVar()
octal_label = ttk.Label(root, textvariable=octal_result)
octal_label.pack()

hexadecimal_result = tk.StringVar()
hexadecimal_label = ttk.Label(root, textvariable=hexadecimal_result)
hexadecimal_label.pack()

# Run the tkinter main loop
root.mainloop()
```
