```python
import os
from tkinter import *
from PIL import Image, ImageTk
from tkinter import filedialog

def load_images(directory):
    image_files = [file for file in os.listdir(directory) if file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))]
    images = []
    for file in image_files:
        image_path = os.path.join(directory, file)
        original_image = Image.open(image_path)
        # Resize the image to fit the window
        resized_image = original_image.resize((700, 700))
        image = ImageTk.PhotoImage(resized_image)
        images.append(image)
    return images

def forward(img_no):
    global label
    global button_forward
    global button_back
    global button_exit
    label.grid_forget()
    label = Label(image=image_list[img_no-1])
    label.grid(row=1, column=0, columnspan=3)
    button_back = Button(root, text="Back", command=lambda: back(img_no-1), state=ACTIVE if img_no > 1 else DISABLED)
    button_forward = Button(root, text="Forward", command=lambda: forward(img_no+1), state=ACTIVE if img_no < len(image_list) else DISABLED)
    button_back.grid(row=5, column=0)
    button_exit.grid(row=5, column=1)
    button_forward.grid(row=5, column=2)

def back(img_no):
    global label
    global button_forward
    global button_back
    global button_exit
    label.grid_forget()
    label = Label(image=image_list[img_no-1])
    label.grid(row=1, column=0, columnspan=3)
    button_back = Button(root, text="Back", command=lambda: back(img_no-1), state=ACTIVE if img_no > 1 else DISABLED)
    button_forward = Button(root, text="Forward", command=lambda: forward(img_no+1), state=ACTIVE if img_no < len(image_list) else DISABLED)
    button_back.grid(row=5, column=0)
    button_exit.grid(row=5, column=1)
    button_forward.grid(row=5, column=2)

def exit_viewer():
    root.destroy()

root = Tk()
root.title("Image Viewer")
root.geometry("700x750")

# Ask user for the image directory
image_directory = filedialog.askdirectory()

image_list = load_images(image_directory)
label = Label(image=image_list[0])
label.grid(row=1, column=0, columnspan=3)

button_back = Button(root, text="Back", command=lambda: back(0), state=DISABLED)
button_exit = Button(root, text="Exit", command=lambda: exit_viewer())
button_forward = Button(root, text="Forward", command=lambda: forward(1), state=ACTIVE if len(image_list) > 1 else DISABLED)

button_back.grid(row=5, column=0)
button_exit.grid(row=5, column=1)
button_forward.grid(row=5, column=2)

root.mainloop()
```
